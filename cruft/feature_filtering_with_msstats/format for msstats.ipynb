```python
import pandas as pd
import numpy as np
```

```python
#0.2 ng files
peptides = pd.read_csv('data/02ng/Task2-SearchTask/AllQuantifiedPeptides.tsv', sep='\t')
peptides.head()
```

```python
# peptides[peptides['Gene Names'] == 'PGM1']
# peptides[peptides['Gene Names'] == 'RRS1']
peptides[peptides[ 'Gene Names'] == 'ACO2']
```

```python
peptides.columns
```

```python
peptides = peptides[['Gene Names', 'Sequence', 'Base Sequence',
                    'Intensity_Ex_Auto_J3_30umTB_02ngQC_60m_1-calib',
                   'Intensity_Ex_Auto_J3_30umTB_02ngQC_60m_2-calib',
                   'Intensity_Ex_Auto_K13_30umTA_02ngQC_60m_1-calib',
                   'Intensity_Ex_Auto_K13_30umTA_02ngQC_60m_2-calib',
                   'Intensity_Ex_Auto_W17_30umTA_02ngQC_60m_3-calib',
                   'Intensity_Ex_Auto_W17_30umTA_02ngQC_60m_4-calib'
                    ]]

peptides = peptides.rename({'Intensity_Ex_Auto_J3_30umTB_02ngQC_60m_1-calib': 'run1',
                            'Intensity_Ex_Auto_J3_30umTB_02ngQC_60m_2-calib': 'run2',
                            'Intensity_Ex_Auto_K13_30umTA_02ngQC_60m_1-calib': 'run3',
                            'Intensity_Ex_Auto_K13_30umTA_02ngQC_60m_2-calib': 'run4',
                            'Intensity_Ex_Auto_W17_30umTA_02ngQC_60m_3-calib': 'run5',
                            'Intensity_Ex_Auto_W17_30umTA_02ngQC_60m_4-calib': 'run6'
                           }, axis=1)
```

```python
peptides.head()
```

```python
pep_long = pd.melt(peptides, id_vars=['Gene Names','Sequence','Base Sequence'], value_vars=['run1','run2','run3','run4','run5','run6'])
pep_long.rename({'Gene Names':'ProteinName',
                 'Sequence': 'PeptideModifiedSequence',
                 'Base Sequence' :'PeptideSequence',
                 'variable':'Run',
                 'value':'Intensity'}, axis=1, inplace=True)
pep_long.head()
```

```python
#add columns that we don't have
pep_long['PrecursorCharge'] = np.nan
pep_long['FragmentIon'] = np.nan
pep_long['ProductCharge'] = np.nan
pep_long['IsotopeLabelType'] = np.nan
pep_long['Condition'] = 1
pep_long['BioReplicate'] = pep_long['Run']
```
